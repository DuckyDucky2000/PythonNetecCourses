**🧪 Ejercicio — Sistema de tienda con constantes**

🎯 Objetivo
Aplicar el uso de constantes en un caso práctico, entendiendo su importancia para evitar valores “mágicos” en el código.

🧠 **Contexto**

Estás desarrollando un sistema sencillo para una tienda donde el usuario puede ingresar el precio de un producto, y el sistema calcula el total con descuento e IVA.

✏️ **Instrucciones:**

Define las siguientes constantes:

DESCUENTO = 0.10

IVA = 0.16 Solicita al usuario el precio del producto usando input()

Realiza: Aplicar descuento Aplicar IVA Mostrar resultados

📌 **Restricciones**

Usa constantes (no valores directos) Convierte el input a número (float) Muestra resultados claros

SALIDA ESPERADA

Ingresa el precio del producto: 500

--- RESULTADO ---

Precio base: 500.0

Precio con descuento: 450.0

Precio final con IVA: 522.0
